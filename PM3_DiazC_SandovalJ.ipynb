## 2. Modelo que representa el problema

En el curso se cubrió el tema de oscilador armónico. El problema del oscilador armónico tiene un potencial -o bien, una energía potencial- de la forma

$$V(x)=k\frac{x^2}{2}$$

que corresponde a la fuerza de Hooke:

$$F=-\nabla V(x)=-kx$$

A este problema también se le conoce como pozo de potencial, pues si una partícula tiene una energía total $E_0$, ésta podrá oscilar dentro del pozo, mas no salir de él, como se muestra a continuación:


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(0,3,100)
y=4*(x-1.5)**2
E=5+0*x
plt.figure(figsize=(10,10))
plt.plot(x,y,label='$V(x)$')
plt.plot(x,E,'k',label='$E_x$')
plt.xlabel('$x$',fontsize=15)
plt.ylabel('$V(x)$',fontsize=15)
plt.plot([1.5],[0.05],'o')
plt.show()
```


En este caso, trataremos un pozo de doble potencial conocido como el _Oscilador Duffing_. La energía potencial es la siguiente:

$$V(x)=-2x^2+x^4$$

Y su Hamiltoniano, es decir, la energía total del sistema, se encuentra dada por:

$$\mathcal{H_0}(x,p)=E_0=\frac{p^2}{2m}-2x^2+x^4$$

Y tiene la siguiente forma:

```python
x=np.linspace(-2,2,50)
plt.figure(figsize=(8,8))
plt.plot(x,-2*x**2+x**4)
plt.xlabel('$x$',fontsize=15)
plt.ylabel('$V(x)$',fontsize=15)
plt.axhline(y=0, color='k')
plt.grid()
plt.show()
```

Podemos ver que existen dos situaciones en este problema:

+ Cuando la energía $E_0<0$, la partícula está confinada dentro del pozo izquierdo o derecho.
+ Cuando la energía $E_0>0$, la partícula está confinada dentro de todo el potencial, pero puede oscilar libremente dentro de él.
